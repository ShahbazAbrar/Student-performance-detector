# Student Performance Project — Interactive Demo

This notebook walks through the final project outputs so you can demo them in viva.
Run the cells in order. It only reads generated files from the `Output/` folder and does not retrain models.

```python
# 1. Import libraries and set paths
import json
from pathlib import Path
import pandas as pd
from IPython.display import display, Image

PROJECT_ROOT = Path('..').resolve().parents[0] if Path('Output').exists() else Path('.')
OUTPUT_DIR = Path('Output')
MODEL_RESULTS = OUTPUT_DIR / 'model_results.csv'
PREDICTION = OUTPUT_DIR / 'prediction_output.json'
SEARCH = OUTPUT_DIR / 'search_path.json'
CSP = OUTPUT_DIR / 'csp_solution.json'
KNOW = OUTPUT_DIR / 'knowledge_inference.json'
FIG_DIR = OUTPUT_DIR / 'Figures'

print(f'Using Output directory: {OUTPUT_DIR.resolve()}')
```

```python
# 2. Model results (comparison)
if MODEL_RESULTS.exists():
    df = pd.read_csv(MODEL_RESULTS)
    display(df)
else:
    print('Model results file not found:', MODEL_RESULTS)
```

```python
# 3. Prediction sample
if PREDICTION.exists():
    with open(PREDICTION, 'r', encoding='utf-8') as f:
        pred = json.load(f)
    print('Prediction output (sample):')
    print(json.dumps(pred, indent=2))
else:
    print('Prediction output not found:', PREDICTION)
```

```python
# 4. Search agent output (BFS path)
if SEARCH.exists():
    with open(SEARCH, 'r', encoding='utf-8') as f:
        search = json.load(f)
    print('Search agent output:')
    print(json.dumps(search, indent=2))
else:
    print('Search output not found:', SEARCH)
```

```python
# 5. CSP solution
if CSP.exists():
    with open(CSP, 'r', encoding='utf-8') as f:
        csp = json.load(f)
    print('CSP solution:')
    print(json.dumps(csp, indent=2))
else:
    print('CSP output not found:', CSP)
```

```python
# 6. Knowledge base inference
if KNOW.exists():
    with open(KNOW, 'r', encoding='utf-8') as f:
        know = json.load(f)
    # Show key parts: initial facts, rules fired, final recommendation
    print('Initial facts (primary case):')
    display(know.get('initial_facts', {}))
    print('
Rules fired:')
    display(know.get('rules_fired', []))
    print('
Final recommendation:')
    print(know.get('final_recommendation', ''))
else:
    print('Knowledge inference file not found:', KNOW)
```

```python
# 7. Show saved figures from Output/Figures (if any)
import os
if FIG_DIR.exists():
    imgs = sorted([p for p in FIG_DIR.iterdir() if p.suffix.lower() in {'.png', '.jpg', '.jpeg'}])
    if not imgs:
        print('No figures found in', FIG_DIR)
    else:
        for p in imgs:
            print('Figure:', p.name)
            display(Image(filename=str(p)))
else:
    print('Figures directory not found:', FIG_DIR)
```

## How to use this demo during viva
- Run each cell to show outputs and figures.
- Use the printed `prediction`, `search_path`, `csp_solution`, and `knowledge_inference` to narrate the workflow: ML -> Risk -> Search -> CSP -> KB.
- The notebook is read-only for results: it does not retrain models. If you need to re-run training, run `Src/ml_pipeline.py` separately.
